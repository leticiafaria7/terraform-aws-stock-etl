## Imports

```python
import pandas as pd
import datetime

import plotly.express as px
import plotly.io as pio

template_plotly = dict(
    layout = dict(
        plot_bgcolor = 'white',
        xaxis = dict(tickfont = dict(size = 10, color = 'gray')),
        yaxis = dict(tickfont = dict(size = 10, color = 'gray'))
    )
)

pio.templates['template_default'] = template_plotly
pio.templates.default = 'template_default'
```

```python
from os import walk
from typing import List

def listar_arquivos_pasta(folder_path: str) -> List[str]:

    files = []
    for (dirpath, dirnames, filenames) in walk(folder_path):
        files.extend(filenames)

    return files
```

## Tabela dimensão dos ativos

```python
df_setores = pd.read_excel('refined/ativos_ibov.xlsx')
print(f"Tamanho da tabela: {df_setores.shape[0]}")
df_setores.head(10)
```

```python
df_setores[['setor']].value_counts().reset_index()
```

```python
arquivos = listar_arquivos_pasta('scraped/')
df = pd.DataFrame()

for arquivo in arquivos:
    tmp = pd.read_csv(f"scraped/{arquivo}")
    df = pd.concat([df, tmp])

df = df[['date', 'codigo', 'price']]
```

```python
print(f"Horário da execução:    {pd.to_datetime(datetime.datetime.now()).round('s')}")
print(f"Quantidade de arquivos: {len(arquivos)}")
```
